```
import numpy as np
```

```
x = np.array((2,4,6,7,8)) #np.array is a function take iterable and return np.ndarray object
type(x)
print(f"x type: {type(x)}\n{x}")
```

Array Rank: Number of dimensions:

```
x = np.array((
    [1, 8, 5, 9],
    [4, 8, 9, 7],
    [9, 8, 4, 2]
), dtype=np.float64)
print(f"Array x rank: {x.ndim}")

y = np.zeros(
    (2, 3, 4) # passed as an argument represents the shape of the ndarray
)
print(f"Array y rank: {y.ndim}")
```

Using np.ndarray.shape to get a tuple contains the ndarray dimensions

```
print(f"Array x: dimensions: {x.shape}")
print(f"Array y: dimensions: {y.shape}")
```

```
print(f"DType x: {x.dtype}")
print(f"array x size: {x.size}")
```

```
np.save("my_array", x) # save np.ndarray to the my_array.npy

c = np.load('my_array.npy')
print(c)
```

```
xz = np.zeros((2, 3), dtype=int) # 2 number of rows, 3 number of columns
xo = np.ones((2, 3), dtype=np.float64)
xf = np.full((2, 3), 7, dtype=np.float64)
xeye = np.eye(5)
xdiag = np.diag([1, 2, 3, 4, 5])
print(f"xz:\n{xz}")
print(f"xo:\n{xo}")
print(f"xf:\n{xf}")
print(f"xeye:\n{xeye}")
print(f"xdiag:\n{xdiag}")
```
