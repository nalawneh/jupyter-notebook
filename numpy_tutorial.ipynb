```
x = np.array((2,4,6,7,8)) #np.array is a function take iterable and return np.ndarray object
type(x)
print(f"x type: {type(x)}\n{x}")
```

Array Rank: Number of dimensions:

```
x = np.array((
    [1, 8, 5, 9],
    [4, 8, 9, 7],
    [9, 8, 4, 2]
))
print(f"Array x rank: {x.ndim}")

y = np.zeros(
    (2, 3, 4) # passed as an argument represents the shape of the ndarray
)
print(f"Array y rank: {y.ndim}")
```
